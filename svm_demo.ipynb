# 4.Load MNIST

```python
import add_fun
from tensorflow.examples.tutorials.mnist import input_data      # 借助tensorflow下载MNIST数据库
# Download MNIST
mnist = input_data.read_data_sets("MNIST_data/", one_hot=False)

# Training Set
print('********************************')
x_train = mnist.train.images
y_train = mnist.train.labels
print('Shape of x_train: ', x_train.shape)
print('Shape of y_train: ', y_train.shape)

# Testing Set
print('********************************')
x_test = mnist.test.images
y_test = mnist.test.labels
print('Shape of x_test: ', x_test.shape)
print('Shape of y_test: ', y_test.shape)
```

```python
# Visualization
import matplotlib.pyplot as plt
i = 16
im = x_train[i]
im = np.reshape(im, (28,28))
plt.imshow(im, cmap='gray')
plt.show()
print('Label: ', y_train[i])
```

# 5. PCA（主成分分析）

![title](image/weishuzainan.png)

![title](image/pca.png)

```python
from sklearn import svm, metrics
from sklearn.decomposition import PCA

# 参数设置
DR_component = 100      #降维之后的维度，即主成分的个数
svm_kernel = 'rbf'
c = 1      # 优化问题中的惩罚系数，越小--决策平面越平滑， 越大--决策面保证更多的训练数据分类正确
gamma = 0.003      # 
```

```python
# PCA
print('--------------------------------------------------------------------------------------------------')
add_fun.Status_Print('PCA', True)
pca = PCA(n_components=DR_component).fit(x_train)
x_train_pca = pca.transform(x_train)
x_test_pca = pca.transform(x_test)
add_fun.Status_Print('PCA', False)
print('--------------------------------------------------------------------------------------------------')
```

# 6. Linear SVM

![title](image/svml.png)

```python
# Linear SVM
print('--------------------------------------------------------------------------------------------------')
add_fun.Status_Print('Linear SVM', True)
linear_svm = svm.LinearSVC()
linear_svm.fit(x_train, y_train)
expected = y_test
predict = linear_svm.predict(x_test)
print("Classification report for classifier %s:\n%s\n"
    % (linear_svm, metrics.classification_report(expected, predict)))
add_fun.Status_Print('Linear SVM', False)
print('--------------------------------------------------------------------------------------------------')
```

```python
# Linear SVM with PCA
print('--------------------------------------------------------------------------------------------------')
add_fun.Status_Print('PCA + Linear SVM', True)
linear_svm = svm.LinearSVC()
linear_svm.fit(x_train_pca, y_train)
expected = y_test
predict = linear_svm.predict(x_test_pca)
print("Classification report for classifier %s:\n%s\n"
    % (linear_svm, metrics.classification_report(expected, predict)))
add_fun.Status_Print('PCA + Linear SVM', False)
print('--------------------------------------------------------------------------------------------------')
```

# 7. Kernel SVM

![title](image/svmk.png)

![title](image/kernel.png)

```python
# Kernel SVM
print('--------------------------------------------------------------------------------------------------')
add_fun.Status_Print('Kernel SVM', True)
kernel_svm = svm.SVC(probability=False, kernel=svm_kernel, C=c, gamma=gamma)
kernel_svm.fit(x_train, y_train)
expected = y_test
predict = kernel_svm.predict(x_test)
print("Classification report for classifier %s:\n%s\n"
    % (kernel_svm, metrics.classification_report(expected, predict)))
add_fun.Status_Print('Kernel SVM', False)
print('--------------------------------------------------------------------------------------------------')
```

```python
# Kernel SVM with PCA
print('--------------------------------------------------------------------------------------------------')
add_fun.Status_Print('PCA + Kernel SVM', True)
kernel_svm = svm.SVC(probability=False, kernel=svm_kernel, C=c, gamma=gamma)
kernel_svm.fit(x_train_pca, y_train)
expected = y_test
predict = kernel_svm.predict(x_test_pca)
print("Classification report for classifier %s:\n%s\n"
    % (kernel_svm, metrics.classification_report(expected, predict)))
add_fun.Status_Print('PCA + Kernel SVM', False)
print('--------------------------------------------------------------------------------------------------')
```
